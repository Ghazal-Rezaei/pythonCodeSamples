Lotka Volterra for juvenile-adult model

```python
from matplotlib import pyplot as plt
from operator import add,truediv
```

```python
j = 15
a = 10
juveniles = [j]
adults = [a]

for i in range(8):
    updated_j=1.6*a
    updated_a=0.3*j+0.8*a
    j=updated_j
    a=updated_a
    juveniles.append(j)
    adults.append(a)
```

```python
plt.gca().xaxis.grid(True)
plt.gca().yaxis.grid(True)
plt.plot(juveniles, label='juveniles')  
plt.xlabel('year')
plt.ylabel('population')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.gca().xaxis.grid(True)
plt.gca().yaxis.grid(True)
plt.plot(adults, label='adults') 
plt.xlabel('year')
plt.ylabel('population')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.gca().xaxis.grid(True)
plt.gca().yaxis.grid(True)
plt.plot(list(map(add,juveniles,adults)) , label='total population') 
plt.xlabel('year')
plt.ylabel('population')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.gca().xaxis.grid(True)
plt.gca().yaxis.grid(True)
plt.plot(list(map(truediv,juveniles,adults)) , label='juvenile/adult') 
plt.xlabel('year')
plt.ylabel('ratio')
plt.legend(loc='upper left')
plt.show()
```
